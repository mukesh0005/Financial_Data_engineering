```python
from yahooquery import Ticker

# Fetch data for Google (Alphabet Inc. - GOOGL)
googl = Ticker("GOOGL")

# Get summary details
summary = googl.summary_detail
print(summary)

# Get asset profile (Company Info)
profile = googl.asset_profile
print(profile)

# Get financial data
financials = googl.financial_data
print(financials)
```
